```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import requests
%matplotlib inline

df = pd.read_csv('sample.csv', sep='\n')
Std = df.std(ddof=1)
Mean = df.mean()
Median = df.median()
fig, axes = plt.subplots(dpi=350)

hist = sns.histplot(data=df,x='selection', kde=True)
hist.axes.set_title("Распределение данных", fontsize=14)
hist.set_xlabel("Показатели", fontsize=12)
hist.set_ylabel("Количество", fontsize=12)

axes.plot([Mean, Mean], [0, 138],c='orange',marker=".")
axes.text(Mean, 105, 'Mean')

axes.plot([Std, Std], [0, 155],c='g',marker=".")
axes.text(Std, 120, '1σ')

axes.plot([Std*2, Std*2], [0, 200],c='b',marker=".")
axes.text(Std*2, 145, '2σ')

plt.show()
```
