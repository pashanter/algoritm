```python
!pip install plotly
import plotly
import plotly.graph_objs as go
import plotly.express as px
from plotly.subplots import make_subplots

import numpy as np
import pandas as pd

x = np.arange(-3, 3, 0.001)
def f(x):
  y = []
  for i in x:
   if i > 0:
     y.append(10**i)
   elif i == 0:
     y.append(0)
   elif i < 0:
     y.append(-10**(-i))
  return y

fig1 = go.Figure()
fig1.update_yaxes(range=[-1000, 1000], zeroline=True, zerolinewidth=2, zerolinecolor='LightPink')
fig1.update_xaxes(range=[-3, 3], zeroline=True, zerolinewidth=2, zerolinecolor='#008000')
fig1.add_trace(go.Scatter(x=x, y=f(x), marker=dict(color='#0099ff'), name='f(x)'))
fig1.update_layout(legend_orientation="h",
                  legend=dict(x=.1, xanchor="right"),
                  title="Part 1",
                  xaxis_title="x Axis Title",
                  yaxis_title="y Axis Title")

fig2 = go.Figure()
fig2.update_yaxes(range=[-3, 3], type='log', zeroline=True, zerolinewidth=2, zerolinecolor='LightPink')
fig2.update_xaxes(range=[-3, 3], zeroline=True, zerolinewidth=2, zerolinecolor='#008000')
fig2.add_trace(go.Scatter(x=x, y=f(x), marker=dict(color='#0099ff'), name='f(x)'))
fig2.update_layout(legend_orientation="h",
                  legend=dict(x=.1, xanchor="right"),
                  title="Part 2",
                  xaxis_title="x Axis Title",
                  yaxis_title="y Axis Title")

fig3 = go.Figure()
fig3.update_yaxes(range=[-3, 3], zeroline=True, zerolinewidth=2, zerolinecolor='LightPink')
fig3.update_xaxes(range=[-3, 3], zeroline=True, zerolinewidth=2, zerolinecolor='#008000')
fig3.add_trace(go.Scatter(x=x, y=x, marker=dict(color='#0099ff'), name='f(x)', hovertemplate='<b>y</b>: 10<sup>%{y}</sup><br><b>x</b>: %{x}'))
fig3.update_layout(legend_orientation="h",
                  legend=dict(x=.1, xanchor="right"),
                  title="Part 3",
                  xaxis_title="x Axis Title",
                  yaxis_title="y Axis Title")
fig3.update_layout(
    yaxis = dict(
        tickvals = [-3, -2, -1, 0, 1, 2, 3],
        ticktext = ['10<sup>-3</sup>', '10<sup>-2</sup>', '10<sup>-1</sup>', '10<sup>0</sup>', '10<sup>1</sup>', '10<sup>2</sup>', '10<sup>3</sup>']
    )
)

fig1.show()
fig2.show()
fig3.show()
```
