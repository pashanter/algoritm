```python
import plotly.graph_objs as go
from numpy import *
import pandas as pd


df = pd.read_csv('Распределение_по_возрастам.csv',
                 sep=';', index_col='Unnamed: 0')
x = df.columns
num_steps = len(x)
fig = go.Figure(data=[go.Scatter(x=df.index, y=df[x[-1]],
                         mode='markers', name='Численность населения РФ по возрастным группам', marker=dict(size=16, color=sign(df[x[-1]] - df[x[-2]]), colorscale=[[0, "red"], [1, "green"]]))])

frames = []
frames.append(go.Frame(name=str(0), data=[go.Scatter(x=df.index, y=df[x[0]],
                         mode='markers', name='Численность населения РФ по возрастным группам', marker=dict(size=16, color='black'))]))
for i in range(1, len(x)):
    frames.append(go.Frame(name=str(i),data=[go.Scatter(x=df.index, y=df[x[i]], mode='markers', name='Численность населения РФ по возрастным группам', marker=dict(size=16, color=sign(df[x[i]] - df[x[i - 1]]), colorscale=[[0, "red"], [1, "green"]]))]))

steps = []
for i in range(num_steps):
    step = dict(
        method = 'animate',
        args = [[str(i)]],
        label=x[i]
    )
    steps.append(step)
sliders = [dict(
    currentvalue = {"suffix": " год", "font": {"size": 20}},
    steps = steps,
    active = len(x) - 1
)]

fig.update_traces(
    hoverinfo="all", hovertemplate="Возрастная группа: %{x}<br>Численность группы: %{y}")
fig.update_yaxes(range=[0, df.max().max() + 2000])
fig.update_layout(title='Численность населения РФ по возрастным группам', xaxis_title='Возрастная группа',
                  yaxis_title='Численность населения', updatemenus=[dict(direction="left",
                                    pad = {"r": 10, "t": 80},
                      x = 0,
                      xanchor = "right",
                      y = 0.1,
                      yanchor = "top",
                      showactive=False,
                      type="buttons",
                      buttons=[dict(label="►", method="animate", args=[None, {"fromcurrent": True}]),
                                             dict(label="❚❚", method="animate", args=[[None], {"frame": {"duration": 0, "redraw": False},
                                                                                               "mode": "immediate",
                                                                                               "transition": {"duration": 0}}])])],
                  )

fig.layout.sliders = sliders
fig.frames = frames
fig.show()
```
